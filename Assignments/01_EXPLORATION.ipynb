```python
import geopandas as gpd
import pandas as pd
```

```python
squirrels = pd.read_csv(
    "../Data/2018_Central_Park_Squirrel_Census_-_Squirrel_Data_20240716.csv"
)
```

```python
squirrels.columns
```

```python
squirrels_gdf = gpd.GeoDataFrame(
    squirrels, geometry=gpd.points_from_xy(squirrels.X, squirrels.Y)
)
```

```python
squirrels_gdf.plot()
```
